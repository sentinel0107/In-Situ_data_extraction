```python
import glob
import pandas as pd
```

```python
filenames = []

count_1 = 0
count_2 = 0
```

```python
for filepath in glob.iglob('/home/jovyan/CR5000_raintab/*.dat'):
    filenames.append(filepath)
    # print(filenames)
    count_1 += 1
for file in filenames:
    print(file)
    count_2 += 1
    df = pd.read_csv(file,skiprows=1,engine='python')
    year_list = df['TIMESTAMP']
    rn_list = df['Rain_mm_Tot']
    df_2 = pd.DataFrame({'year': year_list,'rainfall': rn_list})
    save_name = file.split('/')
    #print(save_name)
    new_name = save_name[4].split('.')
    #print(new_name[0])
    df_2.to_csv('/home/jovyan/CR5000_raintab/output/'+new_name[0]+'.csv')
```
