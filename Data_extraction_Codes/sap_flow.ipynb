```python
import glob
import pandas as pd
```

```python
# make a list of filenames located in a folder

filenames = []

for filepath in glob.iglob('SAP_old/*.dat'):
    filenames.append(filepath)
    
print(len(filenames))
```

```python
for file in filenames:
    print(file)
    if file == 'SAP_old/CR23X_soil_20200218.dat':
        continue
    else:
        data = open(file).readlines()

        filtered_data = []
        for lines in data:
            if lines[0:3] == '131':
                filtered_data.append(lines)
            else:
                continue

        rn_list = []
        year_list = []
        doy_list= []
        time = []
        c1 = []
        c2 = []
        c3 = []

        for numbers in filtered_data:
            x = numbers.split(',')
            rn_list.append(int(x[0]))
            year_list.append(int(x[1]))
            doy_list.append(int(x[2]))
            time.append(int(x[3]))
            c1.append(float(x[10]))
            c2.append(float(x[23]))
            y = x[24].split('\n')
            c3.append(float(y[0]))

        df = pd.DataFrame(
            {'Record_No.': rn_list,
             'Year': year_list,
             'Day of Year': doy_list,
             'Time': time,
             'SFD1': c1,
             'SFD2': c2,
             'SFD3': c3
            }
        )

        save_name = file.split('/')
        new_name = save_name[1].split('.')
        df.to_csv('sap_csv/'+new_name[0]+'.csv')
```
